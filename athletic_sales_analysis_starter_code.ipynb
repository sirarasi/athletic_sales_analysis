### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
athletic_sales_2020_df = pd.read_csv('Resources/athletic_sales_2020.csv')
athletic_sales_2021_df = pd.read_csv('Resources/athletic_sales_2021.csv')
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
joined_sales_2020_2021_rows = pd.concat([athletic_sales_2020_df, athletic_sales_2021_df], axis="rows", join="inner")
#joined_sales_2020_2021_rows.head(30)
joined_sales_2020_2021_rows = joined_sales_2020_2021_rows.reset_index(drop=True)
joined_sales_2020_2021_rows.head(30)
#joined_sales_2020_2021_rows.tail()
```

```python
# Convert the "invoice_date" to a datetime datatype
joined_sales_2020_2021_rows["invoice_date"] = pd.to_datetime(joined_sales_2020_2021_rows["invoice_date"],dayfirst=False, format="%m/%d/%y",errors="coerce")
#joined_sales_2020_2021_rows["invoice_date"] = pd.to_datetime(joined_sales_2020_2021_rows["invoice_date"],dayfirst=False,format="mixed")
#joined_sales_2020_2021_rows["invoice_date"] = pd.to_datetime(joined_sales_2020_2021_rows["invoice_date"],errors="coerce")
joined_sales_2020_2021_rows["invoice_date"].dtype
joined_sales_2020_2021_rows["invoice_date"].head(5)
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
groupby_region_state_city_products_sold = joined_sales_2020_2021_rows.groupby(['region', 'state', 'city'])[['units_sold']].agg('sum')
               
groupby_Total_Products_Sold_Column = groupby_region_state_city_products_sold.rename(columns={"units_sold": "Total_Products_Sold"})                                

Groupby_Top_Regions = groupby_Total_Products_Sold_Column.sort_values(by=["Total_Products_Sold"], ascending=False)

# Show the top 5 results.
Groupby_Top_Regions.head(5)
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
region_state_city_products_sold = pd.pivot_table(joined_sales_2020_2021_rows, 
                                 index=['region', 'state', 'city'],
                                 values='units_sold',
                                 aggfunc='sum')

# Rename the "units_sold" column to "Total_Products_Sold"

Total_Products_Sold_Column = region_state_city_products_sold.rename(columns={"units_sold": "Total_Products_Sold"})

# Sort the renamed pivot table in ascending order to show the top five regions,
# including the state and city that have the greatest number of products sold .
Top_Regions = Total_Products_Sold_Column.sort_values(by=["Total_Products_Sold"], ascending=False)

# Show the top 5 results.
Top_Regions.head(5)
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
groupby_region_state_city_sales = joined_sales_2020_2021_rows.groupby(['region', 'state', 'city'])[['total_sales']].agg('sum')

# Rename the "total_sales" column to "Total Sales"
groupby_Total_Sales_Column = groupby_region_state_city_sales.rename(columns={"total_sales": "Total Sales"})

# Sort the results in ascending order to show the top five regions,
# including the state and city that generated the most sales.
groupby_Top_Regions_Sales = groupby_Total_Sales_Column.sort_values(by=["Total Sales"], ascending=False)

# Show the top 5 results.
groupby_Top_Regions_Sales.head(5)

```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
region_state_city_sales = pd.pivot_table(joined_sales_2020_2021_rows, 
                                 index=['region', 'state', 'city'],
                                 values='total_sales',
                                 aggfunc='sum')

# Optional: Rename the "total_sales" column to "Total Sales"
Total_Sales_Column = region_state_city_sales.rename(columns={"total_sales": "Total Sales"})

# Sort the results in ascending order to show the top five regions,
# including the state and city that generated the most sales.
Top_Regions_Sales = Total_Sales_Column.sort_values(by=["Total Sales"], ascending=False)

# Show the top 5 results.
Top_Regions_Sales.head(5)

```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
Womens_Athletic_Footwear_only = joined_sales_2020_2021_rows.loc[(joined_sales_2020_2021_rows["product"]=="Women's Athletic Footwear")]
Womens_Athletic_Footwear_only.head(5)
```

#### Using `groupby`

```python
#Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
groupby_Units_Women_Athletic_Footwear_sales = Womens_Athletic_Footwear_only.groupby(['retailer','region', 'state', 'city'])[['units_sold']].agg('sum')

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
groupby_Total_WAF_Unit_Column = groupby_Units_Women_Athletic_Footwear_sales.rename(columns={"units_sold": "Womens_Footwear_Units_Sold"})

# Sort the results in ascending order to show the top five retailers along with their,
# region, state and city that sold the most women's athletic footwear.
groupby_Top_WAF_Unit_Sales = groupby_Total_WAF_Unit_Column.sort_values(by=["Womens_Footwear_Units_Sold"], ascending=False)

# Show the top 5 results.
groupby_Top_WAF_Unit_Sales.head(5)
```

#### Using `pivot_table`

```python
#Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
Units_Women_Athletic_Footwear_sales = pd.pivot_table(Womens_Athletic_Footwear_only, 
                                 index=['retailer', 'region', 'state', 'city'],
                                 values='units_sold',
                                 aggfunc='sum')

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
Total_WAF_Unit_Column = Units_Women_Athletic_Footwear_sales.rename(columns={"units_sold": "Womens_Footwear_Units_Sold"})

# Sort the results in ascending order to show the top five retailers along with their,
# region, state and city that sold the most women's athletic footwear.
Top_WAF_Unit_Sales = Total_WAF_Unit_Column.sort_values(by=["Womens_Footwear_Units_Sold"], ascending=False)

# Show the top 5 results.
Top_WAF_Unit_Sales.head(5)
```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
Invoice_Women_Athletic_Footwear_sales = pd.pivot_table(Womens_Athletic_Footwear_only, 
                                 index=['invoice_date'],
                                 values='total_sales',
                                 aggfunc='sum')

# Optional: Rename the "total_sales" column to "Total Sales"
Total_WAF_Invoice_Sales_Column = Invoice_Women_Athletic_Footwear_sales.rename(columns={"total_sales": "Total Sales"})

# Show the table.
Total_WAF_Invoice_Sales_Column
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.

Total_sales_foreachday = Total_WAF_Invoice_Sales_Column.resample('D').sum()

# Sort the resampled pivot table in ascending order on "Total Sales".
Top_WAF_resample_Day_Sales = Total_sales_foreachday.sort_values(by=["Total Sales"], ascending=False)

# Show the top 10 results.
Top_WAF_resample_Day_Sales.head(10)
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
Total_sales_foreachweek = Total_WAF_Invoice_Sales_Column.resample('W').sum()

# Sort the resampled pivot table in ascending order on "Total Sales".
Top_WAF_resample_Week_Sales = Total_sales_foreachweek.sort_values(by=["Total Sales"], ascending=False)

# Show the top 10 results.
Top_WAF_resample_Week_Sales.head(10)
```
